```python
# ------------------------------------------------------------
# Move ONLY raw landing files from old Volume to new Volume
# (Bronze tables are UC-managed, so we do NOT copy a bronze/ folder)
# ------------------------------------------------------------

# Old Volume base path where your raw files currently live
old_base = "/Volumes/workspace/default/olist_stage"  # Existing volume location (workspace catalog)

# New Volume base path you want under the olist catalog
new_base = "/Volumes/olist/stage/olist_stage"        # New volume location (olist catalog)

# Create the new volume base folder (safe even if it already exists)
dbutils.fs.mkdirs(new_base)                          # Ensures /Volumes/olist/stage/olist_stage exists

# Create the raw landing directory inside the new volume
dbutils.fs.mkdirs(f"{new_base}/olist_raw")           # Ensures /Volumes/olist/stage/olist_stage/olist_raw exists

# Copy the raw files folder from old volume to new volume (recursive copy)
# - Third argument True means "recurse" (copy all files/subfolders)
dbutils.fs.cp(f"{old_base}/olist_raw", f"{new_base}/olist_raw", True)

# List the new raw folder to confirm files copied successfully
display(dbutils.fs.ls(f"{new_base}/olist_raw"))
```
